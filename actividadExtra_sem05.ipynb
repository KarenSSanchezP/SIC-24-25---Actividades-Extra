## **¿Qué es Big O?**

La notación Big O es una representación relativa de la complejidad de un algoritmo, nos dice la cantidad de operaciones que necesita hacer un algoritmo, es con esto que podemos comparar diferentes algoritmos siempre y cuando sean del mismo tipo (algoritmos de búsqueda, en este caso), además, generalmente nos muestra el peor de los casos.

### Búsqueda Binaria

El algoritmo de búsqueda binaria funciona en matrices ordenadas, este algoritmo elige un elemento central del arreglo o matriz y compara el valor buscado con el elemento central, si es igual, retorna este valor, si es menor reduce el espacio de búsqueda a la mitad inferior del arreglo, si es mayor reduce el espacio de búsqueda a la mitad superior del arreglo, realiza esto sucesivamente hasta encontrar el valor buscado y reducir el espacio a un arreglo de longitud 1.

En el mejor de los casos, solo  realizará una comparación para encontrar el valor buscado, mientras que en el peor de los casos al ir reduciendo el espacio de búsqueda cada vez más, tendremos: **N, N/2, N/2², N/2³, ..., N/2^k**
El espacio de búsqueda final debe ser igual a 1, entonces: N/2^k = 1, aplicamos logaritmo y tendremos que una complejidad (Big O) de: ***O(log n)***

### Bubble Sort

El algoritmo de burbujas es un algoritmo de ordenamiento sencillo que realiza comparaciones de elementos de par en par, revisa el elemento actual con el siguiente y los intercambia de posición si están desordenados. Este proceso se repite hasta que la lista esté completamente ordenada.

En el mejor de los casos, cuando la lista ya está ordenada, el algoritmo realiza una sola revisión para confirmar que no hay necesidad de hacer intercambios, entonces tendremos una complejidad de **O(n)**. En el peor de los casos, que es cuando la lista está en un orden inverso al requerido, el algoritmo realizará n - 1 revisiones y en cada revisión compara múltiples elementos, entonces tendremos una complejidad de **O(n²)**.

### Selection Sort
El Selection Sort selecciona repetidamente el elemento más pequeño (o más grande, según el caso) del subarreglo no ordenado y lo coloca en su posición correcta al inicio del arreglo.

Este algoritmo siempre tiene que buscar el elemento con el valor mínimo en la lista, por lo que la complejidad en el mejor caso y en el peor caso sigue siendo de **O(n²)**. Esto es independientemente del orden inicial de la lista.

### Insertion Sort 

Toma un elemento de la lista desordenada y lo inserta en la posición correcta dentro de la parte ordenada. Este proceso se repite hasta que toda la lista esté ordenada.

En el mejor de los casos, cuando la lista ya está ordenada el algoritmo realiza **n - 1 comparaciones**, solo para confirmar el orden, lo que resulta en una complejidad de **O(n)**. Mientras que en el peor de los casos, cuando la lista está ordenada de manera inversa, se requiere mover cada elemento completamente a través de la lista ordenada, lo que lleva a una complejidad **O(n²)**.

### Merge Sort 

Es un algoritmo que tiene la filosofía de divide y vencerás. El proceso que realiza es dividir el arreglo en mitades recursivamente, hasta que cada subarreglo tenga un solo elemento, luego combina (fusiona) los subarreglos en orden.

En el mejor de los casos y el peor de los casos, la cantidad de divisiones y fusiones necesarias es siempre la misma, ya que el algoritmo opera recursivamente dividiendo el arreglo en mitades, esto da una complejidad de **O(n log n)** en ambos casos.

### Quick Sort 

Es un algoritmo que también utiliza la filosofía de divide y vencerás, trabaja eligiendo un elemento como pivote y particionando el arreglo en dos subarreglos, uno con elementos menores y otro con elementos mayores al pivote.

En el mejor de los casos, si el pivote divide el arreglo en partes aproximadamente iguales en cada iteración, la complejidad es **O(n log n)**. Mientra que en el peor de los casos, si el pivote es el elemento más pequeño o más grande en cada iteración, el algoritmo se degrada a una complejidad de **O(n²)**.

### **Cuadro Comparativo:** ***Big O de cada algoritmo, Mejor caso vs. Peor caso***

| **Algoritmo**       | **Big O (Mejor Caso)** | **Big O (Peor Caso)** |
|---------------------|------------------------|-----------------------|
| Búsqueda Binaria    | O(1)                   | O(log n)              |
| Bubble Sort         | O(n)                   | O(n²)                 |
| Selection Sort      | O(n²)                  | O(n²)                 |
| Insertion Sort      | O(n)                   | O(n²)                 |
| Merge Sort          | O(n * log n)           | O(n * log n)          |
| Quick Sort          | O(n * log n)           | O(n²)                 |
